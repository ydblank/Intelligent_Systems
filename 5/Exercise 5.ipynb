# A. Nonlinear Transformation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from scipy.stats import linregress
!pip install feature_engine


dataset =  pd.read_csv('china_gdp.csv')
dataset.dropna(inplace=True)
dataset.head()
```

```python
x = dataset['Year']
y = dataset['Value']

slope, intercept, rvalue, pvalue, stderr = linregress(x, y)
y_pred = slope * x + intercept
```

```python
residual = y - y_pred
plt.scatter(x, residual, c='red', label='Residuals')
plt.axhline(y=0, color='blue', linestyle='--', label='Zero Residual Line')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.legend()
plt.title('Residual Plot')
plt.grid(True)
plt.show()
```

```python
r2 = r2_score(y, y_pred)
print(f"R squared: {r2}")
```

```python
y_log = np.log(y)
slope, intercept, rvalue, pvalue, stderr = linregress(x, y_log)

y_pred = np.exp(intercept) * np.exp(slope * x)

# Compute r2 again
r2 = r2_score(y, y_pred)
print(f"R squared [2]: {r2}")
```

```python
residual = y - y_pred

plt.scatter(x, residual, c='red', label='Residuals')
plt.axhline(y=0, color='blue', linestyle='--', label='Zero Residual Line')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.legend()
plt.title('Residual Plot')
plt.grid(True)
plt.show()
```
